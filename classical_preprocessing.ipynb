```python
# pre image processing for classical CNN

from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import os
import numpy as np
from PIL import Image
import cv2
import glob
```

```python
# splitting according to 9:1 ratio
# Set the path to the dataset directory
dataset_path_g = "/home/iisers/Documents/MRI_dataset/glioma_tumor"
dataset_path_m = "/home/iisers/Documents/MRI_dataset/meningioma_tumor"
dataset_path_n = "/home/iisers/Documents/MRI_dataset/no_tumor"
dataset_path_p = "/home/iisers/Documents/MRI_dataset/pituitary_tumor"

# Get a list of all image filenames in the dataset directory
filenames_glicoma = os.listdir(dataset_path_g)
filenames_meningiom = os.listdir(dataset_path_m)
filenames_no_tumor = os.listdir(dataset_path_n)
filenames_pituitary = os.listdir(dataset_path_p)

# Split the filenames into training and testing sets
train_filenames_glicoma, test_filenames_glicoma = train_test_split(filenames_glicoma, test_size=0.1, random_state=42)
train_filenames_meningiom, test_filenames_meningiom = train_test_split(filenames_meningiom, test_size=0.1, random_state=42)
train_filenames_no_tumor, test_filenames_no_tumor = train_test_split(filenames_no_tumor, test_size=0.1, random_state=42)
train_filenames_pituitary, test_filenames_pituitary = train_test_split(filenames_pituitary, test_size=0.1, random_state=42)

# Print the number of images in each set 
print(f"Number of images in training set in glicoma: {len(train_filenames_glicoma)}")
print(f"Number of images in testing set in glicoma: {len(test_filenames_glicoma)}")

print(f"Number of images in training set in meningiom: {len(train_filenames_meningiom)}")
print(f"Number of images in testing set in meningiom: {len(test_filenames_meningiom)}")

print(f"Number of images in training set in no tumor: {len(train_filenames_no_tumor)}")
print(f"Number of images in testing set in no tumor: {len(test_filenames_no_tumor)}")

print(f"Number of images in training set in pituitary: {len(train_filenames_pituitary)}")
print(f"Number of images in testing set in pituitary: {len(test_filenames_pituitary)}")
```

```python
#read an image and resize it to (128, 128,1)
# Set input and output directories
input_dir = "/home/iisers/Documents/MRI_dataset/glioma_tumor"
output_dir = "/home/iisers/Documents/MRI_dataset/glioma_tumor"

# Set target size
target_size = (128, 128)

# Get list of image files in input directory
files = glob.glob(os.path.join(input_dir, '*.jpg'))

# Loop over files in batches and resize
batch_size = 100
for i in range(0, len(files), batch_size):
    batch_files = files[i:i+batch_size]
    batch_images = []
    for file in batch_files:
        image = cv2.imread(file)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        image = cv2.resize(image, target_size)
        batch_images.append(image.reshape((*target_size, 1)))
    batch_images = np.array(batch_images)
    output_path = os.path.join(output_dir, f'batch_{i}.npy')
    np.save(output_path, batch_images)
```

```python
patch_size = (2,2) # patch size is the size of the image patch that will be extracted from the image. 64 means 64*64 image patch
stride = 1 # stride is the number of pixels to shift the patch over the image.
padding = 'same' # padding is used to make sure that the output of the convolutional layer has the same size as the input.
# pad width is the number of pixels to add to the edges of the image. 
# The first two values are the number of pixels to add to the top and bottom of the image, and the second two values are the number of pixels to add to the left and right of the image.
# Load batch file
batch_images

# Loop over images in batch
for i in range(batch_images.shape[0]): # 100 images in batch
    # Load image
    image_array = (batch_images[i])
    image_array = image_array.astype(np.uint8)
    #image_array_1 = batch_images[i]

    # Apply padding if necessary
    if padding == 'same': 
        pad_width = ((1,1),(1,1),(0,0)) # The arguments to the pad_width parameter are a tuple of tuples. The first tuple contains the number of pixels to add to the top and bottom of the image, and the second tuple contains the number of pixels to add to the left and right of the image.
        image_array = np.pad(image_array, pad_width, mode='constant')

    # Create patches using sliding window
    patches = []
    for i in range(0, image_array.shape[0] - patch_size[0] + 1, stride):
        for j in range(0, image_array.shape[1] - patch_size[1] + 1, stride):
            patch = image_array[i:i+patch_size[0], j:j+patch_size[1], :]
            patches.append(patch)

    # Convert patches to numpy array
    patches = np.array(patches)

    # Save patches as individual image files
    for i in range(patches.shape[0]):

        patch = patches[i, :, :, 0]
        
        patch_image = Image.fromarray(patch)

        #patch_image.save(f'image_{i}.png')
```
